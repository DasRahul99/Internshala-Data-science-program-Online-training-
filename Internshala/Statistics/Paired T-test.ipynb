```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_rel
```

### The information given in the dataset is about a typist's work. The typist makes a certain amount of error while typing in a typewriter the he/she was given a computer and makes error their too. Is the error significantly reduced with the change of device ?

```python
data = pd.read_csv('paired t test.csv')
data
```

```python
#calculating t and p value using scipy library
t_statistic,p_value = stats.ttest_rel(data['Errors using typewriter'],data['Errors using a computer'])
p_value
```

### As p value less than 0.05 we can reject the null hypothesis so there is a significance difference after changing device
