```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

## Bigmart sales dataset we need to find out Item_Outlet_Sale( Independent/ Target Variable)

```python
data=pd.read_csv('train.csv')
data
```

### To keep the number of features same in test and train dataset after dummification we need to dummify the entire dataset.

```python
data = pd.get_dummies(data)
```

### As target variable is continuous so it's a regrression problem

```python
# taking first 8000 rows as train abd remaining rows as test dataset
train = data[0:7999]
test = data[8000:]
```

```python
test
```

### We need to readjust the dataset according to sklearn implimentation. sclearn takes two separate arguements for any short of algorithm. It need indipendent variable and target variable separately. As train dataset contains both of them so we need to separate out these.

```python
# creating set of independent variables for train dataset
x_train=train.drop('Item_Outlet_Sales',axis=1) # axis=1 to specify that it dropped from the columns
x_train
```

```python
# creting set of target/dependent variable for train dataset
y_train = train['Item_Outlet_Sales']
```

#### We also need separate arguements for test data set to predict and model checking

```python
x_test = test.drop('Item_Outlet_Sales',axis=1)
observation = test['Item_Outlet_Sales'] # as we need to predict this
```

```python
from sklearn.linear_model import LinearRegression
```

### sklearn first creates an object and then the methods are fited to that object

```python
lreg = LinearRegression()
```

### sklearn needs input in a form of number only i.e. int or float. So we need to create numeric features out of all the categorical or string features. We will be using a concept called dummification where we create dummies of variable. For example consider "Item_Fat_content" variable. Where ever "Low Fat" is occured it will replace it with 1 and other places as 0.

```python
#creating dummy variable for train dataset 
x_train=pd.get_dummies(x_train)
x_train
```

#### x_train has been dummified as number of columns sigficantly increased

```python
#creating dummy for test
x_test = pd.get_dummies(x_test)
```

### Another feature of LinesrRegression is is expects non-missing values. So we have to impute the missing values.

```python
x_train.fillna(0,inplace=True)
x_test.fillna(0,inplace=True)
```

#### LinearRegression has a function fit that fits the training data in the model.

```python
# fitting the model
lreg.fit(x_train,y_train)
```

### Prediction

```python
pred=lreg.predict(x_test)
pred
```

### Model Evaluation

### R-square Test

```python
# performing R-Square on test dataset
lreg.score(x_test,observation) #(independent variables, true value)
```

```python
# performing R-squared test on train dataset
lreg.score(x_train,y_train)
```

### Here we can see train dataset have higher R-squared value so performance dropped in test dataset

### RMSE

```python
from sklearn.metrics import mean_squared_error
```

```python
# RMSE for test dataset
np.sqrt(mean_squared_error(observation,pred)) #(True value,predicted value)
```

```python
# RMSE for train dataset 
np.sqrt(mean_squared_error(y_train,lreg.predict(x_train) ))
```

### Again RMSE pf train is lower so performence dropped for test dataset
